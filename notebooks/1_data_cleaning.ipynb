```python
from datasets import load_dataset
import pandas as pd
#load PuBMedQA
dataset=load_dataset("qiaojin/PubMedQA", "pqa_artificial")
df=dataset["train"].to_pandas()

#Save raw copy
df.to_csv("../data/raw/pubmedqa_raw.csv", index=False)
print("shape:",df.shape)
print("columns:",df.columns.tolist())
df.head()

```

```python
#Check nulls
print("Nulls:\n",df.isnull().sum())
#check duplicate on pubid only
print("\nDuplicate pubids:",df['pubid'].duplicated().sum())
#check label balance
print("\nlabel counts:\n",df['final_decision'].value_counts())
#check context column
print("\nContext type:",type(df['context'][0]))
print("\nContent example:",df['context'][0])
```

```python
df['context_text']=df['context'].apply(lambda x:''.join(x['contexts']))
df['context_labels']=df['context'].apply(lambda x:''.join(x['labels']))
df['context_meshes']=df['context'].apply(lambda x:''.join(x['meshes']))
#check result
print("Content Text:\n",df['context_text'][0])
print("\nContext Labels:\n", df['context_labels'][0])
print("\nContext Meshes:\n",df['context_meshes'][0])
```

```python
import os
os.makedirs('../data/processed', exist_ok=True)
df.to_csv('../data/processed/cleaned_data.csv', index=False)
print("Saved!")
```
