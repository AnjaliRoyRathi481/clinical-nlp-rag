```python
import faiss
import numpy as np
import pandas as pd
from sentence_transformers import SentenceTransformer
```

```python
vectors=np.load("data/processed/embeddings.npy").astype("float32")
print("Vector shape:",vectors.shape)
dim=vectors.shape[1]
index=faiss.IndexFlatL2(dim)
index.add(vectors)
print("Total vectors stored:",index.ntotal)
faiss.write_index(index,"data/processed/faiss_index")
print("Faiss database saved sucessfully")
```

```python
df = pd.read_csv("data/processed/cleaned_data.csv")
index=faiss.read_index("data/processed/faiss_index")
model=SentenceTransformer("all-MiniLM-L6-v2")
 #search_query="What are the symptoms of Diabetes?"
def search(query,top_k=3):
 
 query_vector=model.encode([query]).astype("float32")

 distance,positions=index.search(query_vector,top_K)
 for rank,pos in enumerate(positions[0],1): 
    print(f"Rank{rank}:{df.iloc[pos]['context_text']}")
    print(f"distance:{distance[0][rank-1]:.4f}")
    while True:
     query=input("Ask a cquestion(or type 'quit to stop)")
    if query.lower()=='quit':
      break
    search(query)

```
